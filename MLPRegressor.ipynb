```python
from sklearn.neural_network import MLPRegressor
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt 
import numpy as np
import warnings

warnings.filterwarnings("ignore")
```

```python
workHours = np.array([20, 15, 10, 3])
GPAs = np.array([10, 9, 6, 2])

workHours = workHours.reshape(-1, 1)
GPAs = GPAs.reshape(-1, 1)
```

```python
model = LinearRegression()
model.fit(GPAs, workHours)
```

```python
a = model.coef_
b = model.intercept_
```

```python
plt.plot(GPAs, workHours, 'o')
plt.xlabel('GPAs')
plt.ylabel('workHours')
```

```python
x = [GPAs[0], GPAs[-1]]
y = a * x + b
plt.plot(x, y)
plt.grid(True)
```

```python
net = MLPRegressor(solver = 'lbfgs', 
                   activation = 'tanh',
                   alpha = 1e-5,
                   hidden_layer_sizes=(3, 3),
                   random_state = 1)

net.fit(GPAs, workHours)
```

```python
GPAFiner = np.arange(0, 10.1, .1).reshape(-1, 1)
```

```python
GPAFiner
```

```python
wPredict = net.predict(GPAFiner)
```

```python
wPredict
```

```python
plt.plot(GPAFiner, wPredict, '-m')
```

```python
gpa = np.array(5)
work = net.predict(gpa.reshape(-1, 1))
```

```python
plt.plot (gpa, work, 'dr')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[0],[1],[2]])
t = np.array([[2],[3],[2]])
```

```python
# including bias
x = np.concatenate((x, np.ones([len(x),1])), axis = 1)
```

```python
w = np.array([[1],[2]]) 

alpha = 0.1
maxEpochs = 50
globalError = 10 ** (-3)
```

```python
for index in range(maxEpochs):
    y = np.matmul(x, w)
    e = t - y
    dw = alpha * x * e
    dw = np.sum(dw, axis = 0)
    w = w + dw.reshape(len(w), 1)
    gE = np.sum(e**2)
    print(gE)

    if(gE < globalError):
        break

print('========')
print('weights are', w)

plt.close('all')
plt.plot(x[:, 0], t, 'og')
plt.plot(x[:, 0], y, 'd-r')
plt.grid('true')
```

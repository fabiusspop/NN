```python
import numpy as np
import matplotlib.pyplot as plt

plt.close('all')
```

```python
eps = 10**(-7)
```

```python
def myFunct(x):
    return(np.sin(x) + x**2/50 + x/4-.5)

def myFunctDerivative(x):
    return(np.cos(x) + x/25 + 1/4)
```

```python
def plotDerivative(x):
    dx = 1
    m = myFunctDerivative(x)
    dy = m
    y = myFunct(x)
    plt.plot(x,y, '.r')
    plt.plot([x-dx, x],[y-dy, y])
```

```python
x = np.arange(-10, 10, .1)
y = myFunct(x)
plt.plot(x,y)
plt.grid(True)
```

```python
x0 = 1.5

y0 = myFunct(x0)
plotDerivative(x0)

c=.3
eps =10**(-4)

grad = myFunctDerivative(x0)

while np.abs(grad) > eps:
    x0 = x0 - c*grad
    plotDerivative(x0)
    plt.plot([x0],[0],'dm')
    grad = myFunctDerivative(x0)
```

```python
import numpy as np
```

```python
def myFunct(x):
    return((x**2 - 2)**2)
```

```python
def myFunctGrad(x):
    return(4*x**3-4*x)
```

```python
alpha = 0.0001
```

```python
x = 10 * np.random.rand(1)
myEps = 10**-6
myError = 1
```

```python
index = 1 
while myError > myEps:
    dx = -alpha*myFunctGrad(x)
    x = x + dx 
    myError = abs(myFunct(x))
    print(myError)
    index = index + 1
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor
```

```python
x = np.linspace(0,2 * np.pi, 8)
```

```python
x
```

```python
y = np.sin(x)
```

```python
plt.close()
plt.plot(x, y, 'ro')
```

```python
model = LinearRegression()
model.fit(x.reshape(-1, 1), y)
```

```python
yL = model.coef_ * x + model.intercept_
```

```python
yL
```

```python
plt.plot(x, yL)
```

```python
x = x.reshape(-1, 1)
t = y.reshape(-1, 1)
```

```python
net = MLPRegressor(solver = 'lbfgs',
                   activation = 'logistic',
                   alpha = 1e-5,
                   verbose = 1,
                   hidden_layer_sizes=(3),
                   random_state = 1)

net.fit(x, t)
```

```python
xx = np.arange(0,2 * np.pi, .05).reshape(-1, 1)
yP = net.predict(xx)
```

```python
plt.plot(xx, yP, '.-m')
```
